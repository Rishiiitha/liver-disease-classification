```python
df1=pd.read_csv('liver_dataset_smote.csv')
df1.columns
```

```python
df1.drop("Source", axis=1)
for col in df1.columns:
    plt.figure(figsize=(14,4))
    sns.countplot(x=col, hue="Dataset", data=df1, palette="Set2")
    plt.title("column-wise Liver Disease Distribution")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.legend(title="Target", labels=["No Disease (0)", "Disease (1)"])
    plt.show()
```
